```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel(r'C:\Users\asmit\OneDrive\Desktop\Data Analytics\HW2\urldata.xlsx')
scaled_data = data.iloc[:,1:65]
scaled_data = StandardScaler().fit_transform(scaled_data)
```

```python
#PCA with 95% variance
pca1 = PCA(0.95)
pca1.fit(scaled_data)
x_pca1 = pca1.transform(scaled_data)
cov_matx = np.cov(x_pca1.T)
print ( "Minimum number of principal components that maintain 95% of variance  = ", pca1.n_components_, '\n')
eig_vals1, eig_vecs1 = np.linalg.eig(cov_matx)
eig1 = sorted(eig_vals1, reverse = True)
print(eig1)
plt.plot(eig1)
plt.xlabel('Number of components')
plt.ylabel('Eigen values')
plt.title('Screeplot for 0.95 variance')
plt.show()
```

```python
#PCA with 99% variance
pca2 = PCA(0.99)
pca2.fit(scaled_data)
x_pca2 = pca2.transform(scaled_data)
cov_matx = np.cov(x_pca2.T)
print ( "Minimum number of principal components that maintain 99% of variance  = ", pca1.n_components_, '\n')
eig_vals2, eig_vecs2 = np.linalg.eig(cov_matx)
eig2 = sorted(eig_vals2, reverse = True)
print(eig2)
plt.plot(eig2)
plt.xlabel('Number of components')
plt.ylabel('Eigen values')
plt.title('Screeplot for 0.99 variance')
plt.show()
```
